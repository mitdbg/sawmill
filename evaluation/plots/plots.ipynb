```python
import matplotlib.pyplot as plt
import numpy as np


def gen_pie_plot(
    row_var,
    col_var,
    row_labels,
    col_labels,
    data,
    truth,
    filename="plot",
):

    scaled_data = [1-(abs(a) / b) if a != None else None for a, b in zip(data, truth)]

    colors = ["lightgrey"] * len(data)
    for i in range(len(data)):
        high_proximity = 0.9 <= scaled_data[i]
        med_proximity = 0.75 <= scaled_data[i]
        colors[i] = (
            "#9ACD32"
            if high_proximity
            else ("#FFA500" if med_proximity else "#FF6347")
        )

    # Create a 3x3 grid of subplots
    fig, axs = plt.subplots(3, 3, figsize=(10, 10))

    # Iterate over each subplot in the grid
    for i, ax in enumerate(axs.flat):
        # Calculate the color fractions for the pie chart
        white_frac = 1.0 - scaled_data[i]
        colors_i = ['white', colors[i]]
        explode = (0, 0.1) if colors[i] == 'orange' else (0.1, 0)

        # Create the pie chart
        wedges, _ = ax.pie([white_frac, scaled_data[i]], colors=colors_i, startangle=90, explode=explode)

        # Set the title for each subplot
        row = i // 3
        col = i % 3
        #ax.set_title(f'{row_labels[row]} - {col_labels[col]}')

        # Add the value of the colored portion in the middle of the pie chart
        colored_value = int(scaled_data[i] * 100)
        ax.text(0, 0, f'{colored_value}%', ha='center', va='center', fontsize=20, color='black')

        # Set aspect ratio to 'equal' for each subplot
        ax.set_aspect('equal')

    # Add row labels on the left side of each row
    for i, label in enumerate(row_labels):
        axs[i, 0].annotate(
            label,
            xy=(0, 0.5),
            xytext=(-20, 0),
            ha="right",
            va="center",
            xycoords="axes fraction",
            textcoords="offset points",
            fontsize=20,
        )

    # Add column labels on the bottom of each column
    for j, label in enumerate(col_labels):
        axs[0, j].annotate(
            label,
            xy=(0.5, -2.1),
            xytext=(0, 10),
            ha="center",
            va="center",
            xycoords="axes fraction",
            textcoords="offset points",
            fontsize=20,
        )
        
    # Add a horizontal label for the entire plot
    fig.text(0.5, 0.02, col_var, ha="center", va="center", fontsize=20)

    # Add a vertical label for the entire plot
    fig.text(
        0, 0.5, row_var, ha="center", va="center", rotation="vertical", fontsize=20
    )

    # Adjust the spacing between subplots
    #plt.tight_layout()
    plt.subplots_adjust(hspace=0, wspace=0)

    # Display the plot
    plt.show

    plt.savefig(filename + ".jpg", dpi=300, bbox_inches="tight")
```

```python
import matplotlib.pyplot as plt
import numpy as np


def gen_plot(
    row_var,
    col_var,
    row_labels,
    col_labels,
    data,
    positions,
    truth,
    num_bars=5,
    filename="plot",
    var_name="", 
    qualitative=False,
    data2 = [], 
    positions2 = None, 
    truth2 = None,
    var_name2 = None
):
    # Create a 3x3 grid of subplots
    fig, axs = plt.subplots(len(row_labels), len(col_labels), figsize=(10, 10))

    data = [a if i >=0 else None for a,i in zip(data, positions)]
    scaled_data = [a / b if a != None else None for a, b in zip(data, truth)]
    if len(data2) >0:
        data2 = [a if i >=0 else None for a,i in zip(data2, positions2)]
        scaled_data2 = [a / b if a != None else None for a, b in zip(data2, truth2)]

    edge = (
            max(1, max(scaled_data, key=lambda x: float("-inf") if x is None else x))
            * 1.05
        )
    if qualitative:
        scaled_data = [edge if a != None else None for a in scaled_data]
    if qualitative and len(data2) >0:
        scaled_data2 = [edge if a != None else None for a in scaled_data2]
    print(data)
    print(truth)
    print(scaled_data)

    # Iterate over each subplot in the grid
    for i, ax in enumerate(axs.flat):
        
        # Create an array of indices for the bars
        indices = np.arange(num_bars)

        # Create an array of colors for the bars
        colors = ["lightgrey"] * num_bars
        if scaled_data[i] != None:
            high_proximity = 0.8 <= scaled_data[i] <= 1.6 #Edited
            med_proximity = 0.50 <= scaled_data[i] <= 2
            colors[positions[i]] = (
                "#7F9ABA" if qualitative else
                ("#9ACD32" if high_proximity 
                else ("#FFA500" if med_proximity else "#FF6347")
            ))
        if len(data2) >0 and scaled_data2[i] != None:
            high_proximity = 0.66 <= scaled_data2[i] <= 1.5
            med_proximity = 0.50 <= scaled_data2[i] <= 2
            colors[positions2[i]] = (
                "#40E0D0" if qualitative else
                ("#9ACD32" if high_proximity 
                else ("#FFA500" if med_proximity else "#FF6347")
            ))

        # Create the horizontal bars
        l = [edge] * num_bars
        if scaled_data[i] != None:
            l[positions[i]] = scaled_data[i]
        if len(data2) >0 and scaled_data2[i] != None:
            l[positions2[i]] = scaled_data2[i]
        bars = ax.barh(indices, l, color=colors)


        # add the text label
        if scaled_data[i] != None:
            label = f"{var_name}" if qualitative else f"{var_name}{bars[positions[i]].get_width():.2f}"
            ax.text(
                0.05 * edge,
                bars[positions[i]].get_y() + bars[positions[i]].get_height() / 2,
                label,
                ha="left",
                va="center",
                fontsize=20,
            )
        if len(data2) >0 and scaled_data2[i] != None:
            label = f"{var_name2}" if qualitative else f"{var_name2}{bars[positions2[i]].get_width():.2f}"
            ax.text(
                0.05 * edge,
                bars[positions2[i]].get_y() + bars[positions2[i]].get_height() / 2,
                label,
                ha="left",
                va="center",
                fontsize=20,
            )
        

        # Add a vertical line
        if not qualitative:
            ax.axvline(x=1, color="black", linestyle="--")

        # Invert the y-axis
        ax.invert_yaxis()

        # Set the x-axis range
        ax.set_xlim([min(edge, 0), max(edge, 0)])

        # Hide the axis and the outside outline
        ax.get_yaxis().set_visible(False)
        ax.get_xaxis().set_visible(False)
        ax.spines["top"].set_visible(False)
        ax.spines["left"].set_visible(False)
        ax.spines["right"].set_visible(False)
        ax.spines["bottom"].set_visible(False)

    # Add row labels on the left side of each row
    for i, label in enumerate(row_labels):
        axs[i, 0].annotate(
            label,
            xy=(0, 0.5),
            xytext=(-20, 0),
            ha="right",
            va="center",
            xycoords="axes fraction",
            textcoords="offset points",
            fontsize=20,
        )

    # Add column labels on the bottom of each column
    for j, label in enumerate(col_labels):
        axs[0, j].annotate(
            label,
            xy=(0.5, -2.9),
            xytext=(0, 10),
            ha="center",
            va="center",
            xycoords="axes fraction",
            textcoords="offset points",
            fontsize=20,
        )

    # Add a horizontal label for the entire plot
    fig.text(0.5, -0.02, col_var, ha="center", va="center", fontsize=20)

    # Add a vertical label for the entire plot
    fig.text(
        0, 0.5, row_var, ha="center", va="center", rotation="vertical", fontsize=20
    )

    # Adjust the spacing between subplots
    plt.subplots_adjust(hspace=0.3, wspace=0.3)
    plt.show

    plt.savefig(filename + ".jpg", dpi=300, bbox_inches="tight")
```

```python
import pandas as pd
import matplotlib.pyplot as plt



def stackedplot(parse_time, prep_time, xaxis, xaxis_label, filename, log=False, xmin=None, xmax=None, ymin=None, ymax=None, add_legend=True):
    plt.rcParams.update({'font.size': 20})
    df = pd.DataFrame(data)

    # Create the stacked line plot with colored areas
    plt.figure(figsize=(4,5))

    plt.plot([], [], label='Parsing Time', color='lightgrey', alpha=1)
    a1= plt.fill_between(xaxis, parse_time, color='lightgrey', alpha=1)

    plt.plot([], [], label='Var. Gen. Time', color="#7F9ABA", alpha=1)
    a2 = plt.fill_between(xaxis, parse_time, prep_time, color="#7F9ABA", alpha=1)


    # Customize plot labels and title
    plt.xlabel(xaxis_label)
    if add_legend:
        plt.ylabel('Time (s)')

    if log:
        plt.yscale('log')
        plt.xscale('log')
    
    if xmin != None:
        plt.xlim(left=xmin)
    if xmax != None:
        plt.xlim(right=xmax)
    if ymin != None:
        plt.ylim(bottom=ymin)
    if ymax != None:
        plt.ylim(top=ymax)

    # Add a legend with a custom title and position
    if add_legend:
        plt.legend([a2, a1], ['         ', '     '], loc='upper left')

    # Show the plot
    plt.tight_layout()
    plt.savefig(filename)
```

```python
import matplotlib.pyplot as plt
import networkx as nx





def draw_graph_v2(filename, row_labels, row_var, col_labels, col_var, var_names, found_cause, found_conf):

    # Define different edge colors for each subplot
    cause_colors = ["#9ACD32" if x else "#FF6347" for x in found_cause]
    conf_colors = ["#9ACD32" if x else "#FF6347" for x in found_conf]

    # Define a function to draw the graph with specified edge colors
    def draw_graph_with_colors(i, ax):
        G = nx.DiGraph()
        nodes = [var_names['cause'], var_names['outcome'], var_names['confounder']]
        edges = [(var_names['cause'], var_names['outcome']), (var_names['confounder'], var_names['outcome']), 
                 (var_names['confounder'], var_names['cause'])]
        node_positions = {var_names['cause']: (0.3, 0.3), var_names['outcome']: (0.7, 0.3), var_names['confounder']: (0.5, 0.7)}
        G.add_nodes_from(nodes)
        G.add_edges_from(edges)

        edge_colors = [cause_colors[i], conf_colors[i], conf_colors[i]]

        nx.set_node_attributes(G, node_positions, 'pos')
        pos = nx.get_node_attributes(G, 'pos')
        node_size = 1000
        nx.draw(G, pos, ax=ax, node_color='lightgrey', node_size=node_size, with_labels=True,  font_size=20, edge_color = edge_colors)
        #nx.draw_networkx_edge_labels(G, pos, font_size=8, font_color=edge_colors, ax=ax, label_pos=0.5)
        nx.draw_networkx_edges(G, pos, edgelist=G.edges(), edge_color=edge_colors, width=5, ax=ax, arrows=True, node_size=node_size)

    # Create a 3x3 grid of subplots
    fig, axes = plt.subplots(3, 3, figsize=(10, 10))

    

    # Draw the graphs in each subplot with different edge colors
    for i, ax in enumerate(axes.flat):
        draw_graph_with_colors(i, ax)
        ax.margins(0.2)


    # Add row labels on the left side of each row
    for i, label in enumerate(row_labels):
        axes[i, 0].annotate(
            label,
            xy=(-0.1, 0.5),
            xytext=(-20, 0),
            ha="right",
            va="center",
            xycoords="axes fraction",
            textcoords="offset points",
            fontsize=14,
        )

    # Add column labels on the bottom of each column
    for j, label in enumerate(col_labels):
        axes[0, j].annotate(
            label,
            xy=(0.5, -2.9),
            xytext=(0, 10),
            ha="center",
            va="center",
            xycoords="axes fraction",
            textcoords="offset points",
            fontsize=14,
        )

    # Add a horizontal label for the entire plot
    fig.text(0.5, 0.02, col_var, ha="center", va="center", fontsize=16)

    # Add a vertical label for the entire plot
    fig.text(
        0, 0.5, row_var, ha="center", va="center", rotation="vertical", fontsize=16
    )

    # Adjust the spacing between subplots
    plt.subplots_adjust(hspace=0.3, wspace=0.3)

    # Adjust the layout and spacing
    plt.tight_layout(rect=[0, 0, 1, 0.97])

    

    # Display the plot
    plt.show

    plt.savefig(filename + ".jpg", dpi=300, bbox_inches="tight")


```
